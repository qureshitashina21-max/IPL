```python
season = df['season'].value_counts().sort_index()
```

```python
plt.bar(season.index,season.values)
plt.xlabel("season")
plt.ylabel("Number of matches")
plt.title("Matches played per Season")
plt.show()
```

###insight
The number of IPL matches increased from the early seasons

and peaked around 2012-2013. showing leaague expansion during that period.

After 2014 the number of matches stabilized ,with only small variation each season.

This indicates that IPL reached a consistent tournament structure in later years.

```python
team_counts = df ['team1'].value_counts()
```

```python
plt.bar(team_counts.index,team_counts.values)
plt.xlabel("team1")
plt.ylabel("matches played")
plt.title("Matches played by each team")
plt.xticks(rotation=90)
plt.show()
```

###insight
mumbai Indians, Kings XI Punjab, and Chennai Super Kings have played the highest number of matches, showing they are among the most consistent and long-standing IPL teams.

Teams like Delhi Daredevils, Rajasthan Royals, and Sunrisers Hyderabad also have high match counts, indicating regular participation across seasons.

Newer or short-lived teams such as Gujarat Lions, Pune Warriors, Kochi Tuskers Kerala, and Rising Pune Supergiants have played significantly fewer matches.

This shows that team longevity strongly impacts total matches playedÂ inÂ theÂ IPL.
